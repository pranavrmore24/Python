```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('maharashtra_hospitals_.csv')
df.head()
```

```python
df.describe()
```

```python
df['Total_Ventilators'] = df['Total_Ventilators'].str.replace('Ten', '10')
df['Patient_Capacity_per_day'] = df['Patient_Capacity_per_day'].str.replace('One Hundred', '100')
df['Ratings'] = df['Ratings'].str.replace('Four Point Five', '4.5')
```

```python
df['Total_Ventilators'] = df['Total_Ventilators'].astype('int64')
df['Patient_Capacity_per_day'] = df['Patient_Capacity_per_day'].astype('int64')
```

```python
df['Ratings'] = pd.to_numeric(df['Ratings'], errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# For categorical columns, fill with the mode
departments_mode = df['Departments_Available'].mode()[0]
df['Departments_Available'] = df['Departments_Available'].fillna(departments_mode)

# For numerical columns, fill with the median
for col in ['Doctor_Count', 'Nurse_Count', 'Ratings', 'Total_Ventilators', 'Patient_Capacity_per_day']:
    col_median = df[col].median()
    df[col] = df[col].fillna(col_median)
```

```python
df.duplicated().sum()
```

```python
#cheking outliers at once
for i in df.select_dtypes(include=np.number).columns:
  plt.title(i)
  sns.boxplot(df[i])
  plt.show()
```

```python
#  capping outliers using IQR method from specific columns
outlier_cols = ["Total_Beds", "Doctor_Count", "Patient_Capacity_per_day", "Ratings"]

for col in outlier_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Clip values beyond the bounds
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
#cheking outliers at once
for i in df.select_dtypes(include=np.number).columns:
  plt.title(i)
  sns.boxplot(df[i])
  plt.show()
```

```python
# Categorical Variables
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.countplot(y='Type', data=df, order = df['Type'].value_counts().index)
plt.title('Distribution of Hospital Types')
```

```python
plt.subplot(2, 2, 2)
sns.countplot(y='Specialty', data=df, order = df['Specialty'].value_counts().index)
plt.title('Distribution of Hospital Specialties')
plt.tight_layout()
plt.show()
```

```python
# Numerical Variables
df[['Total_Beds', 'Total_Ventilators', 'Doctor_Count', 'Nurse_Count', 'Patient_Capacity_per_day', 'Ratings']].hist(figsize=(15, 10), bins=15)
plt.suptitle('Histograms of Numerical Variables')
plt.show()
```

```python
# Bivariate Analysis 
plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
sns.scatterplot(x='Total_Beds', y='Doctor_Count', data=df)
plt.title('Total Beds vs. Doctor Count')

plt.subplot(1, 3, 2)
sns.scatterplot(x='Doctor_Count', y='Nurse_Count', data=df)
plt.title('Doctor Count vs. Nurse Count')

plt.subplot(1, 3, 3)
sns.scatterplot(x='Total_Beds', y='Patient_Capacity_per_day', data=df)
plt.title('Total Beds vs. Patient Capacity')
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
# --- Violin Plot: Ratings by Hospital Type ---

plt.figure(figsize=(10, 6))
sns.violinplot(x='Type', y='Ratings', data=df, palette='muted')
plt.title('Distribution of Ratings Across Hospital Types', fontsize=16)
plt.xlabel('Hospital Type', fontsize=12)
plt.ylabel('Ratings', fontsize=12)
plt.show()
```

```python
# --- Stacked Bar Chart: Emergency Services by Hospital Type ---

# Create a cross-tabulation of the two columns
type_emergency = pd.crosstab(df['Type'], df['Emergency_Services'])

# Plot the stacked bar chart
type_emergency.plot(kind='bar', stacked=True, figsize=(10, 7), colormap='viridis')
plt.title('Emergency Services Availability by Hospital Type', fontsize=16)
plt.xlabel('Hospital Type', fontsize=12)
plt.ylabel('Number of Hospitals', fontsize=12)
plt.xticks(rotation=0) # Keeps the x-axis labels horizontal
plt.legend(title='Emergency Services')
plt.show()
```

```python
# --- Bubble Chart: Beds, Doctors, and Patient Capacity ---

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df,
                x='Total_Beds',
                y='Doctor_Count',
                size='Patient_Capacity_per_day',
                hue='Type',  # Color bubbles by hospital type for an extra dimension
                sizes=(50, 1000),  # Set the range of bubble sizes for better visibility
                alpha=0.7, # Use transparency to see overlapping bubbles
                palette='bright')

plt.title('Hospital Analysis: Beds, Doctors, and Patient Capacity', fontsize=16)
plt.xlabel('Total Beds', fontsize=12)
plt.ylabel('Doctor Count', fontsize=12)
plt.legend(title='Hospital Type & Patient Capacity')
plt.grid(True)
plt.show()
```
